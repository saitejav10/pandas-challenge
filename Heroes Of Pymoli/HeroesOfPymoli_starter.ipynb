## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by SN
group_sn = purchase_data.groupby(["SN"])

#Purchase count per user
count_sn = group_sn["Price"].count()

#Average spent per user
avgprice_sn = group_sn["Price"].mean()

#Total amount spent per user
totalamount_sn = group_sn["Price"].sum()

print(count_sn)
print(avgprice_sn)
print(totalamount_sn)
```

```python
#Dataframe
topspender_df = pd.DataFrame({
                            "Purchase Count":count_sn,
                            "Average Purchase Price":avgprice_sn,
                            "Total Purchase Value": totalamount_sn
                            })

#topspender_df
#Sort the total purchase value column in descending order
top_spender = topspender_df.sort_values("Total Purchase Value",ascending=False)

#Format numbers
top_spender["Average Purchase Price"] = top_spender["Average Purchase Price"].map("${:,.2f}".format) 
top_spender["Total Purchase Value"] = top_spender["Total Purchase Value"].map("${:,.2f}".format) 

top_spender.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name, and Item Price columns
retrieve_items = purchase_data[["Item ID","Item Name","Price"]]
#item_id

#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
most_popular = retrieve_items.groupby(["Item ID", "Item Name"])

#Purchase count per item
popular_count = most_popular["Price"].count()

#Average price spent per item
popular_avg = most_popular["Price"].mean()

#Total purchase value spent per item
popular_total = most_popular["Price"].sum()

print(popular_count)
print(popular_avg)
print(popular_total)
```

```python
#Dataframe
items_df = pd.DataFrame({ 
                          "Purchase Count":popular_count,
                          "Item Price":popular_avg,
                          "Total Purchase Value": popular_total,
})

#items_df.head()

#Sort the purchase count column 
popular_items = items_df.sort_values("Purchase Count",ascending=False)

#Formatting
popular_items["Item Price"] = popular_items["Item Price"].map("${:,.2f}".format) 
popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map("${:,.2f}".format) 

popular_items.head()
```
